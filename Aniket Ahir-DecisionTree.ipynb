```python
import pandas as pd
import numpy as np
from sklearn import tree,metrics
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("mushrooms.csv")
data
```

```python
data.isnull().sum()
```

```python
li=[pd.get_dummies(data[['class','cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing','gill-size', 'gill-color','stalk-shape','stalk-root', 'stalk-surface-above-ring','stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color','ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']],drop_first=True)]
df=pd.concat(li,axis=1)
x=df
li2=[pd.get_dummies(data[['class']],drop_first=True)]
y=pd.concat(li2,axis=1)
x.head()
```

```python
y.head()
```

```python
model = tree.DecisionTreeClassifier()
model
```

```python
mtree_model=model.fit(x,y)
mtree_model
```

```python
model.score(x, y)
```

```python
y_pred = mtree_model.predict(x)
y_pred
```

```python
metrics.accuracy_score(y, y_pred)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
mtree_model_2 = model.fit(x_train, y_train)
```

```python
y_train_pred = mtree_model_2.predict(x_train)
print(metrics.accuracy_score(y_train, y_train_pred))
```

```python
y_test_pred = mtree_model_2.predict(x_test)
print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
param_grid = { "criterion"      : ['gini', 'entropy'],    
               "max_features"   : [1, 2 ],             
               "splitter"       : ['best', 'random'],  
               "min_samples_split"  : [2, 3],
               "min_samples_leaf"  : [ 2, 3]
                
                }
```

```python
optimized_tree = GridSearchCV(mtree_model, param_grid, scoring = 'accuracy', cv = 5)
optimized_tree.fit(x_train, y_train)
optimized_tree.best_score_
```

```python
print("Best Params:")
print(optimized_tree.best_params_)
print("Best estimator:")
print(optimized_tree.best_estimator_)
print("Best cv result:")
print(optimized_tree.cv_results_)
```
